```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.ion()
import pandas as pd
import pickle
```

```python
d = pickle.load(open("calibration_constants.pickle", "rb"))
d
```

```python
fig, axes = plt.subplots(8, 2, sharex=True, sharey=True, figsize=(12, 24))
for channel in range(8):
    for ig, gaintype in enumerate(["high", "low"]):
        ax = axes[channel, ig]
        s = d[channel, gaintype]
        ax.errorbar(
            x=np.arange(len(s.mean)),
            y=s.mean,
            yerr=s.sem,
            fmt='.')
        ax.set_xlim(0, 4096)
        ax.set_title("channel:{0} {1}".format(channel, gaintype))
        ax.grid()
```

```python
fig, axes = plt.subplots(8, 2, sharex=True, sharey=True, figsize=(12, 24))
for channel in range(8):
    for ig, gaintype in enumerate(["high", "low"]):
        ax = axes[channel, ig]
        s = d[channel, gaintype]
        ax.plot(s.n)
        ax.set_xlim(0, 4096)
        ax.set_title("channel:{0} {1}".format(channel, gaintype))
        ax.grid()
        ax.set_ylim(0, 1e4)
```
